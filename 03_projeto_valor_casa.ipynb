## **PROJETO 01**

### Criar uma rede neural simples que prevê preço de casas baseada em uma equação linear.

> uma casa com 1 quarto custa 100k

> uma casa com 2 quartos custa 150k

> uma casa com 3 quartos custa 200 k

### Utilizando o tensorflow: qual a predição para uma casa de 7 quartos (utlizando 1000 épocas).

**OBS:**  A função que descreve o comportamento do valor é: y = 50x + 50

```python
# Importando as bibliotecas utilizadas
import tensorflow as tf
import numpy as np
from tensorflow import keras


# cria-se um modelo sequencial com 1 camada oculta

modelo = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

#defini-se a forma como será computada utilizando o otimizador e a função de erro

modelo.compile(optimizer = 'sgd', loss = 'mean_squared_error', metrics='accuracy')

# Inicia-se os dados de entrada para treinamento
x=[1.0, 2.0, 3.0, 4.0]
y=[1.0, 1.5, 2.0, 2.5]

# definição do modelo e da época do treinamento utilizando fit

modelo.fit(x,y,epochs = 1000)

```

```python
# Modelo de predição

print(modelo.predict([7.0]))
```

 De acordo com a equação o valor deveria ser 400k

 Ou seja a predição foi satisfatória.
